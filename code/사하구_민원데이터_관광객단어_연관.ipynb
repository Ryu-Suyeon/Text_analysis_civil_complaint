```python
# 데이터 불러오기
import re

fileDir = "C:/Users/rsy08/OneDrive/바탕 화면/데이터인문세미나/감천민원_내용.txt" ##수정##
# fileDir = "C:/Users/###/###/###/###/감천민원_내용.txt"
fr = open(fileDir, 'r', encoding='utf-8')
contents = fr.readlines()
fr.close()

reviewList = []
for content in contents:
  content = re.sub('[^가-힣]', ' ', content)
  content = re.sub('[\s]+', ' ', content).strip()
  reviewList.append(content)

for eachReview in reviewList:
  print(eachReview)
```

```python
# kiwipiey 패키지 형태소 분석

from kiwipiepy import Kiwi

kiwi = Kiwi()
# 품사태깅된 문장
reviewPOSList = []
# 단어의 타입 확인
wordType = set()

# 원하는 타입 설정
desired_types = {"NNG"} # 명사 지정해서, 명사만 추출

for eachReview in reviewList:
    # 띄어쓰기
    eachReviewSpace = kiwi.space(eachReview)
    # 품사태깅
    posTagout = kiwi.tokenize(eachReviewSpace)
    sentencePOS = ''
    for eachOut in posTagout:
        result = eachOut[0] + "/" + eachOut[1]
        if eachOut[1] in desired_types:  # 원하는 타입의 단어만 선택
            wordType.add(result)
            sentencePOS = sentencePOS + " " + result
    sentencePOS = sentencePOS.strip()
    reviewPOSList.append(sentencePOS)
    print(sentencePOS)

nng_text=[]

for i in range(len(reviewPOSList)):
    nng_text.append(reviewPOSList)
    
print(reviewPOSList)
```

```python
import pandas as pd

df = pd.DataFrame(index=reviewPOSList, columns=None)
df['text'] = reviewPOSList
print(df)
```

```python
import pandas as pd

# 특정 단어
target_word = '관광객/NNG'

# 새로운 열을 추가하여 특정 단어의 출현 빈도를 계산하여 표시
df['word_count_관광객'] = df['text'].apply(lambda x: x.split().count(target_word))
df = df[df['word_count_관광객'] != 0] # 관광객 단어가 포함되지 않은 텍스트가 있는 행은 삭제하고 df 형성

# 결과 확인
print(df)

# 엑셀 파일로 저장
excel_file_path = '민원텍스트_단어빈도(관광객).xlsx'
df.to_excel(excel_file_path, index=False, header=True)
```
